## Linear Vs Non-Linear

```python
#Linear

import numpy as np

# Sample data
x = np.linspace(-5, 5, 100)
y = x**2  # Non-linear relationship

# Simple neural network (1 layer, linear activation)
def linear_network(x, weight, bias):
    return x * weight + bias

# Initial weights and bias
weight = 1.0
bias = 0.0

# Predictions
y_pred = linear_network(x, weight, bias)

# Plotting
import matplotlib.pyplot as plt

plt.plot(x, y, label="True y (x^2)")
plt.plot(x, y_pred, label="Predicted y (linear)")
plt.legend()
plt.show()
```

```python
# Non-Linear

import numpy as np

# Sample data (same as before)
x = np.linspace(-5, 5, 100)
y = x**2

# Simple neural network (2 layers, ReLU activation)
def relu(x):
    return np.maximum(0, x)

def non_linear_network(x, w1, b1, w2, b2):
    hidden = relu(x * w1 + b1)
    return hidden * w2 + b2

# Initial weights and biases
w1 = 1.0
b1 = -2.0
w2 = 1.0
b2 = 0.0

# Predictions
y_pred = non_linear_network(x, w1, b1, w2, b2)

# Plotting
import matplotlib.pyplot as plt

plt.plot(x, y, label="True y (x^2)")
plt.plot(x, y_pred, label="Predicted y (non-linear)")
plt.legend()
plt.show()
```
